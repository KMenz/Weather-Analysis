### Analysis
* We can see that, as one would usually infer, as we get closer to the equator the temperature rises. You can see though that for the same time of year, the Northern Hemisphere is colder. Look at 40 and -40 Latitude for example. There are more points below 40 degrees for the Northern Hemisphere than the Southern. This could be because this data was taken during the Winter for the Northern Hemisphere, where it would be Summer in the Southern. 
* There is no strong relationship between latitude and humidity. However, interestingly enough there is a band of cities that sits at 100% humidity despite the latitude
* There is no strong relationship between latitude and wind speed. There is a group of cities however that is clustered in the Northern Hemisphere between 0 and 5 mph. More cities than reported in the Southern Hemisphere. This may be due to a collection of data resulting in more cities from the Northern Hemisphere as it seems there are more Northern Cities than Southern for each plot. 


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)

```

```python
#Turn City Into DF
cities_df = pd.DataFrame(cities)
cities_df = cities_df.rename(columns = {0: 'City'})
cities_df.head()
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Call URL and Units
base_url = 'http://api.openweathermap.org/data/2.5/weather?'
units = 'imperial'
count = 0

#Set up columns in df for response
cities_df['Latitude'] = ""
cities_df['Longitude'] = ""
cities_df['Cloudiness'] = ""
cities_df['Humidity'] = ""
cities_df['Max Temp'] = ""
cities_df['Wind Speed'] = ""
cities_df['Country'] = ""



#Create full URL
for index, city in cities_df.iterrows():
    count += 1
    
    query_url = base_url + "appid=" + api_key + "&units=" + units + "&q=" + city['City']

#Try and Except in order to add columns and data to cities_df
    try: 
        weather_response = requests.get(query_url)
        city_json = weather_response.json()
        
       
        
        cities_df.set_value(index,'Latitude', int(city_json['coord']['lat']))
        cities_df.set_value(index, 'Longitude', int(city_json['coord']['lon']))
        cities_df.set_value(index, 'Cloudiness', int(city_json['clouds']['all']))
        cities_df.set_value(index, 'Humidity', int(city_json['main']['humidity']))
        cities_df.set_value(index, 'Max Temp', int(city_json['main']['temp_max']))
        cities_df.set_value(index, 'Wind Speed', int(city_json['wind']['speed']))
        cities_df.set_value(index, 'Country', city_json['sys']['country'])
        
#Print log of process        
    except:
        print(f"No data for this city: {city['City']}")
    print(f"This is city#: {count}")
    print(f"This is: {city['City']}" )
    print(f"This is the requested URL: {query_url}")
        
    
  
        

```

```python
#Remove Rows With Empty Cells
cities_df = cities_df[cities_df.Latitude != '']
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Write DF To CSV
cities_df.to_csv(output_data_file, index = False, header = True)

#Display DF
cities_df.head()
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
plt.scatter(cities_df['Latitude'], cities_df['Max Temp'], marker = 'o', facecolors = 'blue', edgecolors = 'black', alpha = .75)

plt.title('City Latitude vs Max Temperature (F)')
plt.xlabel('Latitude')
plt.ylabel('Max Temp (F)')
plt.grid(True)

plt.savefig('output_data/Lat_vs_Temp.png')
```

#### Latitude vs. Humidity Plot

```python
plt.scatter(cities_df['Latitude'], cities_df['Humidity'], marker = 'o', facecolors = 'blue', edgecolors = 'black', alpha = .75)

plt.title('City Latitude vs Humidity')
plt.xlabel('Latitude')
plt.ylabel('Humidity')
plt.grid(True)

plt.savefig('output_data/Lat_vs_Humid.png')
```

#### Latitude vs. Cloudiness Plot

```python
plt.scatter(cities_df['Latitude'], cities_df['Cloudiness'], marker = 'o', facecolors = 'blue', edgecolors = 'black', alpha = .75)

plt.title('City Latitude vs Cloudiness')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness')
plt.grid(True)

plt.savefig('output_data/Lat_vs_Cloud.png')
```
